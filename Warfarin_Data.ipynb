```python
import pandas as pd
import numpy as np
df=pd.read_csv('Warfarin_Data.csv')

# View columns and size of datasets
print(df.columns)
print(df.shape)
```

```python
# Choose the columns we need
df_selected_cols= df[['PharmGKB Subject ID','Age','Gender','Weight (kg)','Height (cm)','Therapeutic Dose of Warfarin','Cyp2C9 genotypes','VKORC1 genotype: -1639 G>A (3673); chr16:31015190; rs9923231; C/T']]
df_selected_cols.head()

# Rename the columns
df_new=df_selected_cols.rename(columns={'PharmGKB Subject ID':'ID','Age':'Age (years)','Gender':'Gender','Weight (kg)':'Weight (kg)','Height (cm)':'Height (cm)','Therapeutic Dose of Warfarin':'Dose (mg)','Cyp2C9 genotypes':'CYP2C9','VKORC1 genotype: -1639 G>A (3673); chr16:31015190; rs9923231; C/T':'VKORC1 (rs9923231; C/T)'})
df_new.head()

# df_new is a new name for dataframe
```

```python
print("Data Information:")
df_new.info()
print("\n" + "="*50 + "\n")
```

```python
# To remove the whitespace
df_new['Gender']=df_new['Gender'].str.strip()

# Change the letters to lowercase to maintain the consistency
df_new['Gender']=df_new['Gender'].str.lower()

# Change the first letter to uppercase while the rest in lowercase
df_new['Gender']=df_new['Gender'].str.capitalize()

df_new.head()
```

```python
# To export new subset into csv file
file_name_csv = 'Warfarin_new.csv'
df_new.to_csv(file_name_csv, index=False)
print(f"Subset of dataframe has been export to '{file_name_csv}'")
```
